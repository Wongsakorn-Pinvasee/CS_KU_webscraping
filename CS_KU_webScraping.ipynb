```
import urllib
from bs4 import BeautifulSoup
import pandas as pd
```

# **Subjects**

```
URL = "https://cs.sci.ku.ac.th/curriculum/bachelor/y2560/course/"
```

```
page = urllib.request.urlopen(URL)
```

```
soup = BeautifulSoup(page,'html.parser')
```

```
print(soup)
```

```
subjects = soup.find('tbody',attrs={'class':'row-hover'})
```

```
print(subjects)
```

```
subjects.text.strip()
```

```
sj_lists = subjects.text.strip().split("01418")
print(sj_lists[0:5])
```

```
subjects_list = []
for s in sj_lists[1:]:
    a = []
    a.append('01418'+s[0:3])
    a.append(s[3:s.index('(')])
    a.append(s[s.index('(')+1:s.index(')')])
    a.append(s[s.index(')')+1])
    subjects_list.append(a)
```

```
print(subjects_list[0:5])
```

```
subjects_df = pd.DataFrame(subjects_list)
subjects_df.head()
```

```
subjects_df = subjects_df.rename(columns={0:"s_id", 1:'s_name_th', 2:'s_name_eng', 3:'credit'})
subjects_df.head()
```

```
subjects_df.to_csv("CS_KU_subjects.csv", header=True, index=False, encoding='utf-8')
```

# **Teacher**

```
URL = "https://cs.sci.ku.ac.th/personnel/academic-2/"
```

```
page = urllib.request.urlopen(URL)
```

```
soup = BeautifulSoup(page, 'html.parser')
print(soup)
```

```
teacher = soup.findAll('ul', attrs={"class":"elementor-icon-list-items"})
```

```
print(teacher)
```

```
teacher[1].text.strip()
```

```
t_list = []
for t in teacher:
    text = t.text.strip().replace(" AT ","@").replace("อ.","")
    t_list.append(text.split())
```

```
t_list[0:2]
```

```
teacher_df = pd.DataFrame(t_list)
```

```
teacher_df.head()
```

```
teacher_df[teacher_df[8].notnull()]
```

```
teacher_df = teacher_df.rename(columns={0:'t_id', 1:'th_firstname', 2:'th_lastname', 3:'eng_firstname', 4:'eng_lastname', 6:'room', 7:'email', 8:'website'})
```

```
teacher_df.head()
```

```
teachers_df = teacher_df.drop(columns=[5,'website'])
```

```
teachers_df.head()
```

```
teachers_fn_df =  teachers_df.drop([teachers_df.index[len(teachers_df)-2], teachers_df.index[len(teachers_df)-1]])
```

```
teachers_fn_df.tail()
```

```
teachers_fn_df['full_th_name'] = teachers_fn_df['th_firstname'] +  " " +teachers_fn_df['th_lastname']
teachers_fn_df['full_eng_name'] = teachers_fn_df['eng_firstname'] + " "+ teachers_fn_df['eng_lastname']
```

```
teachers_fn_df.head()
```

```
teachers_fn_df = teachers_fn_df.drop(columns=['th_firstname','th_lastname','eng_firstname','eng_lastname'])
```

```
teachers_fn_df.head()
```

```
teachers_fn_df.to_csv("CS_KU_teachers.csv", header=True, index=False, encoding='utf-8')
```
